```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as ureq
```

```python
url = ureq('https://www.medifee.com/hospitals-in-coimbatore')
page = url.read()
page = bs(page , 'html.parser')
url.close()
```

```python
l = ['Bypass Surgery' , 'X-Ray','MRI Scan','CT Scan','Kidney Stone','Hernia Surgery']
```

```python
'''a = page.findAll('a')
print(len(a))

a1 = page.findAll('table',{'class':"table table-bordered table-striped"})
a1'''
a2 = page.findAll('td')
#print(a2[1].a)
ll = []
for i in range(0,len(a2),2):
    a2[i] =  a2[i].a.get('href')
    a2[i] = 'https://www.medifee.com' + a2[i]
    ll.append(a2[i])

for i in ll:
    print(i)
    
    
fo = open('covai.csv','w')
fo.write("Name,City,Address,Bypass Surgery,X-Ray,MRI Scan,CT Scan,Kidney Stone,Hernia Surgery\n")
```

```python
for i in ll:
    url = ureq(i)
    page = url.read()
    page = bs(page,'html.parser')
    url.close()
    
    #Find Name
    h1 = page.findAll('h1')
    print(h1[0].text)
    Name = h1[0].text
    
    #Find address
    ad = page.findAll('div',{'class':"col-lg-6"})
    print(ad[0].p.text)
    ad = ad[0].p.text.split(', ')
    print(ad[-2])
    city = ad[-2]
    add = ''
    for i in ad:
        add = add + i +'/'
    add = add.replace('-','/')    
    print(add)
    
    #Find Feature
    table = page.findAll('tr' )
    f = {}
    s = ''
    for i in table[1::]:
        i = i.text
        i = i.split('Rs. ')
        print(i)

        for j in i[1]:

            if j == '.' :
                break
            else :
                s = s+ j

        i[1] = s

        f[ i[0] ] = i[1]
        s = ''

    for i,j in f.items():
        print(i," : ",j)

        #Extract Feature
        for i,j in f.items():
            print(i," : ",j)
        
    by , xr , ms ,cs , ks , hs = '','','','','',''
    
    for i in f:
        if i in l:
            print(i)
            if i == 'Bypass Surgery':
                by = 'Rs.'+f[i]

            elif i =='X-Ray':
                xr = 'Rs.'+f[i]

            elif i =='MRI Scan':
                ms = 'Rs.'+f[i]

            elif i == 'CT Scan':
                cs = 'Rs.'+f[i]

            elif i == 'Kidney Stone':
                ks = 'Rs.'+f[i]

            elif i == 'Hernia Surgery':
                hs = 'Rs.'+f[i] 
            print(f[i])
            
    fo.write(Name+','+city+','+add+','+by+','+xr+','+ms+','+cs+','+ks+','+hs+'\n')
        
    
    
    
```
